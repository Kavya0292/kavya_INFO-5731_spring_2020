```
#1.1.4 Average word length ---- Method 1
d = open("file.txt", "r", encoding='cp1252')
total_wordcount_sum = 0
total_wordlength_sum = 0
for line in d:
    word_1 = line.split()
    # sum up the word counts
    total_wordcount_sum += len(word_1)
    for w in word_1:
        # sum up the word lengths
        total_wordlength_sum += len(w) 
total_wordlength_average = total_wordlength_sum/float(total_wordcount_sum) # invoke floating point division
print("Average word length:- ", total_wordlength_average)

```

```
#1.1.4 Average word length ---- Method 2
with  open("file.txt", "r", encoding='cp1252') as f:
  w = [len(word) for line in f for word in line.rstrip().split(" ")]
  w_avg = sum(w)/len(w)
  print("Average word length:- ", w_avg)
```

```
import collections
import nltk
from nltk.util import ngrams
d = open("file.txt", "r", encoding='cp1252')
f = d.read()
lo=[]
grams_2=[]
grams_3=[]
file = word_tokenize(f)
for line in file:
    gr=ngrams(nltk.word_tokenize(line),1)
    gr_2=ngrams(nltk.word_tokenize(line),2)
    gr_3=ngrams(nltk.word_tokenize(line),3)
    for i in gr:
      lo.append(i)
    for j in gr_2:
      grams_2.append(j)
    for k in gr_3:
      grams_3.append(k)
top=collections.Counter(lo)
print("1-gram",top.most_common(10))
top_2=collections.Counter(grams_2)
print("2-grams",top_2.most_common(10))
top_3=collections.Counter(grams_3)
print("3-grams",top_3.most_common(5))
  

```
